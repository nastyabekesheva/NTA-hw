```python
def rp2(alpha, beta, p):
    x = [1]
    a = [0]
    b = [0]
    n = p - 1 
    i = 0
    while True:
        if 0 <= x[i] < p / 3:
            a.append(a[i])
            b.append((b[i] + 1) % n)
            print(f"x{i} in S1")
        elif p / 3 <= x[i] < 2 * p /3:
            a.append(2 * a[i] % n)
            b.append(2 * b[i] % n)
            print(f"x{i} in S2")
        elif 2 * p /3 <= x[i] < p:
            a.append((a[i] + 1) % n)
            b.append(b[i])
            print(f"x{i} in S3")
        i += 1
        #print(pow(alpha, a[i], n), pow(beta, b[i], n))
        x.append(pow(alpha, a[i], p) * pow(beta, b[i], p) % p)
        print(f"x{i} = {x[i]}")

        if i % 2 == 0:
            if x[i] == x[i//2]:
                r = (b[i//2] - b[i]) % n
                if r == 0 :
                    print("error")
                    break
                else:
                    a = a[i]-a[i//2]
                    d = math.gcd(r, n)
                    print("d = ", d)
                    if  d != 1:
                        r = r // d
                        n = n // d
                        a = a // d
                    x = pow(r, -1, n) * (a) % n
                    print("x = ", x)
                    while x < p:
                        if pow(alpha, x, p) == beta:
                            print("x = ", x)
                            return x
                        x += n

                    break
            
```

```python
import math
            
def rp3(alpha, beta, p, n):
    x = [1]
    a = [0]
    b = [0]
    i = 0
    while True:
        if x[i] % 3 == 1:
            a.append(a[i])
            b.append((b[i] + 1) % n)
            print(f"x{i} in S1")
        elif x[i] % 3 == 0:
            a.append(2 * a[i] % n)
            b.append(2 * b[i] % n)
            print(f"x{i} in S2")
        elif x[i] % 3 == 2:
            a.append((a[i] + 1) % n)
            b.append(b[i])
            print(f"x{i} in S3")
        i += 1
        x.append(pow(alpha, a[i], p) * pow(beta, b[i], p) % p)
        print(f"x{i} = {x[i]}")

        if i % 2 == 0:
            if x[i] == x[i//2]:
                r = (b[i//2] - b[i]) % n
                if r == 0 :
                    print("error")
                    break
                else:
                    a = a[i]-a[i//2]
                    d = math.gcd(r, n)
                    if  d != 1:
                        r = r // d
                        n = n // d
                        a = a // d
                    x = pow(r, -1, n) * (a) % n
                    print("x = ", x)
                    while x < p:
                        if pow(alpha, x, p) == beta:
                            print("solution x = ", x)
                            return x
                        x += n

                    break
```

```python
def bsgs(alpha, beta, p, n):
    m = int(math.sqrt(n)) + 1
    b = pow(alpha, m, p)

    giant_step = []
    baby_step = []

    for i in range(1, m + 1):
        giant_step.append(pow(b, i, p))
        baby_step.append(beta * pow(alpha, i, p) % p)

    print("GS: ", giant_step)
    print("BS: ", baby_step)

    u, v = 0, 0
    for index1, item1 in enumerate(giant_step):
        if item1 in baby_step:
            index2 = baby_step.index(item1)
            u =  index1 + 1
            v = index2 + 1
            
    print(f"u = {u}, v = {v}")
    x = (m * u - v) % n

    return x
            
start_time = time.time()
print("x =", bsgs(64, 122, 607, 101))
end_time = time.time()
print(f"Elapsed: {end_time - start_time} seconds")            
```
